```python
import pandas
import matplotlib.pyplot as plt
```

```python
# geting the csv file containing the colors
color_df = pandas.read_csv('colors.csv')
color_df
```

```python
# getting total number of unique colors
color_df['name'].nunique()
```

```python
# getting the number of transperent colors
color_df.is_trans.value_counts()
```

```python
# reading the sets dataset
sets_df = pandas.read_csv('sets.csv')
sets_df
```

```python
# getting the oldest theme
sets_df.sort_values('year')
```

```python
# the number different products  LEGO company sold in their first year of operation
sets_df[sets_df['year']==1949]
```

```python
# the top 5 lego set with the most number of parts
sets_df.sort_values('num_parts', ascending=False)
```

```python
sets_by_year = sets_df.groupby('year').count()

num_parts_by_year = sets_by_year['num_parts']

num_parts_by_year_count = num_parts_by_year.count()
sets_by_year.tail()
```

```python
# Group by year and count the number of parts per year
sets_by_year = sets_df.groupby('year').count()

# Ploting the data
plt.figure(figsize=(16,10))
# excluding the last  years
plt.plot(sets_by_year.index[:-1], sets_by_year['num_parts'][:-1])
plt.xlabel('Year')
plt.ylabel('Number of Parts')
plt.show()

```

```python
themes_by_year= sets_df.groupby('year').agg({'theme_id': pandas.Series.nunique})
themes_by_year.tail()
```

```python
# plotting a graph of theme against year
plt.figure(figsize=(16,10))
# excluding the last  years
plt.plot(themes_by_year.index[:-1], themes_by_year['theme_id'][:-1])
plt.xlabel('Year')
plt.ylabel('themes')
plt.show()
```

```python
# ploting theme and number of parts aginst year on the same graph
plt.figure(figsize=(10,6))
ax1 = plt.gca() 
ax2 = ax1.twinx()
ax1.plot(sets_by_year.index[:-1], sets_by_year['num_parts'][:-1], color ='green')
ax2.plot(themes_by_year.index[:-1], themes_by_year['theme_id'][:-1])
plt.xlabel('Year')
ax1.set_ylabel('Number of sets', color ='green')
ax2.set_ylabel('Number of theme', color ='blue')

plt.show()
```

```python
# average number of parts per LEGO set in that year
parts_per_set = sets_df.groupby('year').agg({'num_parts': pandas.Series.mean})
parts_per_set
```

```python
plt.scatter(parts_per_set.index[:-1], parts_per_set.num_parts[:-1])
plt.xlabel('Year')
plt.ylabel('number of parts')
```
