```python
import pandas
import plotly.express as px
```

```python
df_apps =pandas.read_csv('apps.csv')
df_apps.shape
```

```python
# return any random row
df_apps.sample()
```

```python
# dropping 
df_apps.drop(['Last_Updated', 'Android_Ver'], axis=1, inplace = True)
df_apps
```

```python
# looking for rows with NaN 
nan_rows = df_apps[df_apps.Rating.isna()]
nan_rows
```

```python
# dropping all rows with Nan
df_apps_clean = df_apps.dropna()
df_apps_clean.shape
```

```python
#  Finding and Removing Duplicates
duplicated_rows = df_apps_clean[df_apps_clean.duplicated()]
print(duplicated_rows.shape)
duplicated_rows.head()
```

```python
df_apps_clean = df_apps_clean.drop_duplicates(subset=['App', 'Type', 'Price'])
df_apps_clean[df_apps_clean.App == 'Twitter']
```

```python
# inspecting dimension of cleaned dataframe
df_apps_clean.shape
```

```python
# highest rates apps
df_apps_clean.sort_values('Rating', ascending=False).head()
```

```python
# largest Android apps
df_apps_clean.sort_values('Size_MBs', ascending=False).head()
```

```python
# top 50 apps with the most reviews
df_apps_clean.sort_values('Reviews', ascending=False).head(5)
```

```python
# counting the number of occurrences of each rating
ratings = df_apps_clean.Content_Rating.value_counts()
ratings
```

```python
# creating a chart content rating
fig = px.pie(labels=ratings.index,
values=ratings.values,
title="Content Rating",
names=ratings.index,
)
fig.update_traces(textposition='outside', textinfo='percent+label')
```

```python
# creating a doughtnuts chart content rating
fig = px.pie(labels=ratings.index,
values=ratings.values,
title="Content Rating",
names=ratings.index,
hole=0.5,
)
fig.update_traces(textposition='inside', textfont_size=15, textinfo='percent')

fig.show()
```

```python
# checking data type of of the installs coloumn
df_apps_clean['Installs'].describe()
```

```python
# checking data type of the dataset using .info()
df_apps_clean.info()
```

```python
# removing commas frommthe install coloumn and changing the data type to int
df_apps_clean.Installs = df_apps_clean.Installs.astype(str).str.replace(',', "")
df_apps_clean.Installs = pandas.to_numeric(df_apps_clean.Installs)
df_apps_clean[['App', 'Installs']].groupby('Installs').count()
```

```python
# removing commas frommthe install coloumn and changing the data type to int
df_apps_clean.Price = df_apps_clean.Price.astype(str).str.replace('$', "")
df_apps_clean.Price = pandas.to_numeric(df_apps_clean.Price)
df_apps_clean[['App', 'Price']].groupby('Price').count()
# getting the top 20 most expensive apps
df_apps_clean.sort_values('Price', ascending=False).head(20)
```

```python
# eliminating  apps with price greater than 250
df_apps_clean = df_apps_clean[df_apps_clean['Price'] < 250]
df_apps_clean.sort_values('Price', ascending=False).head(10)
```

```python
# highest grossing paid apps
df_apps_clean['Revenue_Estimate'] = df_apps_clean.Installs.mul(df_apps_clean.Price)
df_apps_clean.sort_values('Revenue_Estimate', ascending=False)[:10]
```

```python
# finding  the number of different categories 

df_apps_clean.Category.nunique()
```

```python
# calculating the number of apps per category 

top10_category = df_apps_clean.Category.value_counts()[:10]
top10_category
```

```python
# plotting a bar chart for categories
bar = px.bar(x = top10_category.index,
             y = top10_category.values)

bar.show()
```

```python
 # how often apps are downloaded in a category by grouping all  apps by category and sum the number of installations
category_installs = df_apps_clean.groupby('Category').agg({'Installs': pandas.Series.sum})
category_installs.sort_values('Installs', ascending=True, inplace=True)

# horizontal bar chart by adding the orientation parameter:
h_bar = px.bar(x = category_installs.Installs,
               y = category_installs.index,
               orientation='h',
               title='Category Popularity')

h_bar.update_layout(xaxis_title='Number of Downloads', yaxis_title='Category')
h_bar.show()
```

```python
# Creating  a scatter plot with Plotly

# working out the number of apps in each category .

cat_number = df_apps_clean.groupby('Category').agg({'App': pandas.Series.count})

# combine the two DataFrames.

cat_merged_df = pandas.merge(cat_number, category_installs, on='Category', how="inner")
print(f'The dimensions of the DataFrame are: {cat_merged_df.shape}')
# sorting base on the highest number of installs
cat_merged_df.sort_values('Installs', ascending=False)
```

```python
 scatter = px.scatter(cat_merged_df, # data
                    x='App', # column name
                    y='Installs',
                    title='Category Concentration',
                    size='App',
                    hover_name=cat_merged_df.index,
                    color='Installs')

scatter.update_layout(xaxis_title="Number of Apps (Lower=More Concentrated)",
                      yaxis_title="Installs",
                      yaxis=dict(type='log'))

scatter.show()
```

```python
# counting the number of genres
len(df_apps_clean.Genres.unique())
```

```python
# changing the data type of genres by spliting the strings on the semi-colon and then .stack them.
stack = df_apps_clean.Genres.str.split(';', expand=True).stack()
print(f'We now have a single column with shape: {stack.shape}')
num_genres = stack.value_counts()
print(f'Number of genres: {len(num_genres)}')
```

```python
# bar chart of genres
bar = px.bar(x = num_genres.index[:15], # index = category name
             y = num_genres.values[:15], # count
             title='Top Genres',
             hover_name=num_genres.index[:15],
             color=num_genres.values[:15],
             color_continuous_scale='Agsunset')

bar.update_layout(xaxis_title='Genre',
yaxis_title='Number of Apps',
coloraxis_showscale=False)

bar.show()
```

```python
# verifying the categories with most paid apps
df_free_vs_paid = df_apps_clean.groupby(["Category", "Type"], as_index=False).agg({'App': pandas.Series.count})
df_free_vs_paid
```

```python
# bar chart of Free vs. Paid Apps per Category
g_bar = px.bar(df_free_vs_paid,
               x='Category',
               y='App',
               title='Free vs Paid Apps by Category',
               color='Type',
               barmode='group')

g_bar.update_layout(xaxis_title='Category',
                    yaxis_title='Number of Apps',
                    xaxis={'categoryorder':'total descending'},
                    yaxis=dict(type='log'))

g_bar.show()
```
