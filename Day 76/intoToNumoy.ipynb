```python
# fetching the racoon image int Scipy libiary
img = misc.face()


plt.imshow(img)

plt.title("Raccoon Image")
plt.show()
```

```python
# inspecting properties of racoon
type(img)
```

```python
img.shape
```

```python
img.ndim
```

```python
# converting racoon image to grey scale
sRGB_array = img / 255
grey_vals = np.array([0.2126, 0.7152, 0.0722])
img_gray = sRGB_array @ grey_vals
img_gray = sRGB_array @ grey_vals
img_gray = np.matmul(sRGB_array, grey_vals)
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(np.flip(img_gray), cmap='gray')
```
